### 기존에 하던 Mutation Count가 아니라 TMB로 thresholding 하기 위한 분석
mutation count 323으로 나누면 LUAD만 잘됐음

```python
import pandas as pd

our_index = 'Mutation Count'

# csv_path='/shared/j.jang/pathai/CLAM/dataset_csv/TCGA-lung-LUAD+LUSC-TMB-pan_cancer-323.csv'
csv_path = '/shared/js.yun/CLAM/dataset_csv/TCGA-lung-LUAD+LUSC-TMB-pan_cancer-323.csv'
# slide_data = pd.read_csv(csv_path)['TMB (nonsynonymous)']
slide_data = pd.read_csv(csv_path)[['case_id', 'slide_id', 'label', 'Subtype', 'Mutation Count', 'TMB (nonsynonymous)']]
# print(slide_data)

TMB_distribution = slide_data[our_index].tolist()
# print(TMB_distribution)
```

```python
import matplotlib.pyplot as plt
import numpy as np


###################################
threshold = 200
###################################

def plot_histogram(data, subtype, threshold=323):
    plt.figure(figsize=(7, 5))  # 그래프 크기 조절
    counts, bins, patches = plt.hist(data, bins=25, edgecolor='black')

    # 각 bin의 빈도수 표시
    for count, bin, patch in zip(counts, bins, patches):
        height = patch.get_height()
        plt.annotate(f'{int(count)}', xy=(bin + (bins[1] - bins[0])/2, height),
                     xytext=(0, 3),  # 3 points vertical offset
                     textcoords="offset points",
                     ha='center', va='bottom', fontsize=9)  # 폰트 크기 조절

    # a보다 작은 데이터와 큰 데이터의 개수 계산
    # less_than_threshold = np.sum(np.array(data) < threshold)
    greater_than_threshold = np.sum(np.array(data) >= threshold)
    greater_than_threshold_percent = (greater_than_threshold / len(data)) * 100

    # 최대값, 최소값, a보다 작은/큰 데이터의 개수 표시
    plt.annotate(f'Min: {np.min(data)}', xy=(0.55, 0.90), xycoords='axes fraction', fontsize=12)
    plt.annotate(f'Max: {np.max(data)}', xy=(0.55, 0.85), xycoords='axes fraction', fontsize=12)
    # plt.annotate(f'Count < {threshold}: {less_than_threshold}', xy=(0.6, 0.80), xycoords='axes fraction', fontsize=12)
    plt.annotate(f'Count >= {threshold}: {greater_than_threshold} ({greater_than_threshold_percent:.2f}%)', xy=(0.55, 0.75), xycoords='axes fraction', fontsize=12)

    # x축의 a 위치에 빨간색 세로선 그리기
    plt.axvline(x=threshold, color='red', linestyle='--')

    # 그래프 제목과 라벨 추가
    plt.title(f'TMB Distribution Histogram ({subtype})', fontsize=14)
    # plt.xlabel('TMB (nonsynonymous)', fontsize=12)
    plt.xlabel(our_index, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)

    plt.tight_layout()  # 레이아웃 조절
    plt.show()

# 전체 데이터 histogram
plot_histogram(TMB_distribution, 'LUSC + LUAD', threshold)

# Subtype이 'LUSC'인 데이터만 필터링
LUSC_data = slide_data[slide_data['Subtype'] == 'LUSC'][our_index].tolist()
plot_histogram(LUSC_data, 'LUSC', threshold)

# Subtype이 'LUAD'인 데이터만 필터링
LUAD_data = slide_data[slide_data['Subtype'] == 'LUAD'][our_index].tolist()
plot_histogram(LUAD_data, 'LUAD', threshold)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import _LRScheduler
import math

# Define a simple model
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.linear = nn.Linear(10, 1)
    
    def forward(self, x):
        return self.linear(x)
    
class WarmupCosineLR(_LRScheduler):
    def __init__(self, optimizer, warmup_iter, base_lr, final_lr, total_iters, last_epoch=-1):
        self.warmup_iter = warmup_iter
        self.base_lr = base_lr
        self.final_lr = final_lr
        self.total_iters = total_iters
        super().__init__(optimizer, last_epoch)

    def get_lr(self):
        if self.last_epoch < self.warmup_iter:
            return [base_lr * (self.last_epoch / self.warmup_iter) for base_lr in self.base_lrs]
        else:
            # Cosine annealing
            cos_out = self.final_lr + 0.5 * (self.base_lr - self.final_lr) * (
                1 + math.cos(math.pi * (self.last_epoch - self.warmup_iter) / (self.total_iters - self.warmup_iter)))
            return [cos_out]



# Initialize model
model = SimpleModel()

# Warmup and total iters configuration
warmup_iter = 5
epochs = 3
# total_iters = epochs*20
base_lr = 0.001
final_lr = 0.00001

# Setup optimizer
optimizer = optim.Adam(model.parameters(), lr=base_lr)  # lr is set to 0 because it will be updated by the scheduler

# Dummy data loader (just for demonstration purposes)
dummy_data_loader = [(torch.randn(5, 10), torch.randn(5, 1)) for _ in range(20)]

# Initialize scheduler
scheduler = WarmupCosineLR(optimizer, warmup_iter=warmup_iter, base_lr=base_lr, final_lr=final_lr, total_iters=epochs*len(dummy_data_loader))

# Training loop with learning rate scheduling
for epoch in range(epochs):
    print(f"Epoch {epoch+1}\n-------------------------------")
    for batch_idx, (data, target) in enumerate(dummy_data_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = torch.nn.functional.mse_loss(output, target)
        loss.backward()
        optimizer.step()
        scheduler.step()  # Update learning rate
        
        print(f"Batch {batch_idx+1}, LR: {scheduler.get_lr()[0]:.6f}")

print("Training done!")
```
